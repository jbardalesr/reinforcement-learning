# Week 1: Multi-armed Bandits

Given the estimate update in red, what do you think was the value of the step size parameter we used to update the estimate on each time step?2.
$$Q_{t+1} = Q_t + \alpha_n(R_t - Q_t)$$

```python
def target_estimated(alpha):
    R = [0, 4, 4, 4, 4, 4, 4, 2, 2, 2, 2, 4, 0, 4, 0, 4, 0, 4, 0]
    Q = [0]
    for t in range(len(R)):
        Q.append(Q[t] + alpha(t) * (R[t] - Q[t]))
    plt.title("$Q_{t+1} = Q_t + \\alpha_n(R_t - Q_t)$")
    plt.plot(R, 'o-', label="$R_t$")
    plt.plot(Q, 'o-', c='r', label="$Q_t$")
    plt.xlabel("time")
    plt.legend()
    plt.plot()
```

Question 1: $\alpha_t=1/2$

```python
target_estimated(lambda t: 1 / 2)
```

Question 2: $\alpha_t = 1/t$

```python
target_estimated(lambda t: 1 / t if t != 0 else 1)
```
